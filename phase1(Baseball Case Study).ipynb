# removing outlies

```python
#1st qantile
q1=data.quantile(0.25)
#3rd quantile
q3=data.quantile(0.75)
iqr=q3-q1
```

```python
iqr
```

```python
#same for all deleting outlier
R_high=(q3['R'] +(1.5*iqr['R']))
print(R_high)
a=np.where(data['R']>R_high)
data=data.drop(data.index[a])
print(data.shape)

data.reset_index()
```

```python
#same for all deleting outlier
ERA_high=(q3['ERA'] +(1.5*iqr['ERA']))
print(ERA_high)
a=np.where(data['ERA']>ERA_high)
data=data.drop(data.index[a])
print(data.shape)

data.reset_index()
```

```python
#same for all deleting outlier
SHO_high=(q3['SHO'] +(1.5*iqr['SHO']))
print(SHO_high)
a=np.where(data['SHO']>SHO_high)
data=data.drop(data.index[a])
print(data.shape)

data.reset_index()
```

```python
#same for all deleting outlier
sv_high=(q3['SV'] +(1.5*iqr['SV']))
print(sv_high)
a=np.where(data['SV']>sv_high)
data=data.drop(data.index[a])
print(data.shape)

data.reset_index()
```

```python
#same for all deleting outlier
E_high=(q3['E'] +(1.5*iqr['E']))
print(E_high)
a=np.where(data['E']>E_high)
data=data.drop(data.index[a])
print(data.shape)

data.reset_index()
```

```python
x=data.drop(columns=['W'])
y=data['W']
```

# EDA PROCESSING IS COMPLETED

```python
# data scalling.formula z=(x-mean)/z
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x) 
```

```python
 x_scaled.shape[1]
```

```python
vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(x_scaled,i) for i in range  (x_scaled.shape[1])]
vif['features']=x.columns
```

```python
vif
```

```python
# trainning and spliting data
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.20,random_state=120)
```

```python
from sklearn.linear_model import LinearRegression   # WE SELECT LR because target is continous data
log_reg=LinearRegression()
log_reg.fit(x_train,y_train)
```

```python
# i have selected linear regression bcz here in target column continious  data thats why i have selected linear regression model
```

# lets check how well model fits on train data¶
